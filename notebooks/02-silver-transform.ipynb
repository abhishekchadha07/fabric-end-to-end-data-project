```python
df_sales = spark.table("bronze_sales")
df_customers = spark.table("bronze_customers")
df_products = spark.table("bronze_products")
df_stores = spark.table("bronze_stores")
```

```python
from pyspark.sql.functions import col

df_sales_clean = df_sales \
.withColumn("OrderDate", col("OrderDate").cast("date")) \
.withColumn("Quantity", col("Quantity").cast("int")) \
.withColumn("UnitPrice", col("UnitPrice").cast("double")) \
.withColumn("TotalAmount", col("TotalAmount").cast("double"))
```

```python
df_sales_clean.write.mode("overwrite").saveAsTable("silver_sales")
df_customers.write.mode("overwrite").saveAsTable("silver_customers")
df_products.write.mode("overwrite").saveAsTable("silver_products")
df_stores.write.mode("overwrite").saveAsTable("silver_stores")
```
