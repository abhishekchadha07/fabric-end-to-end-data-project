```python
sales = spark.table("silver_sales")
customers = spark.table("silver_customers")
products = spark.table("silver_products")
stores = spark.table("silver_stores")
```

```python
dim_customer = customers.dropDuplicates(["CustomerID"])
dim_customer.write.mode("overwrite").saveAsTable("gold_dim_customer")
```

```python
dim_product = products.dropDuplicates(["ProductID"])
dim_product.write.mode("overwrite").saveAsTable("gold_dim_product")
```

```python
dim_store = stores.dropDuplicates(["StoreID"])
dim_store.write.mode("overwrite").saveAsTable("gold_dim_store")
```

```python
from pyspark.sql.functions import year,month,dayofmonth,quarter

dim_date = sales.select("OrderDate").dropDuplicates()

dim_date = dim_date \
.withColumn("Year", year("OrderDate")) \
.withColumn("Month", month("OrderDate")) \
.withColumn("Day", dayofmonth("OrderDate")) \
.withColumn("Quarter", quarter("OrderDate"))

dim_date.write.mode("overwrite").saveAsTable("gold_dim_date")
```

```python
fact_sales = sales.select(
"OrderID",
"OrderDate",
"CustomerID",
"ProductID",
"StoreID",
"Quantity",
"UnitPrice",
"TotalAmount"
)

fact_sales.write.mode("overwrite").saveAsTable("gold_fact_sales")
```
